```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
if 'left_str' in df.columns:
    df.drop(columns=['left_str'], inplace=True)


X = df.drop('left', axis=1)
y = df['left']
```

```python
cat_cols = ['department', 'salary']
num_cols = [col for col in X.columns if col not in cat_cols]

# One-hot encode categorical columns
X_cat = pd.get_dummies(X[cat_cols], drop_first=True)
X_num = X[num_cols]

# Combine processed categorical and numerical columns
X_final = pd.concat([X_num, X_cat], axis=1)
```

## Train Test Split

```python
# Stratified split to preserve class distribution
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.2, random_state=123, stratify=y
)

print("Before SMOTE:")
print(y_train.value_counts())

```

```python
smote = SMOTE(random_state=123)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print("After SMOTE:")
print(y_train_smote.value_counts())
```

# 5 fold cross validation and model training
## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import classification_report

# Logistic Regression Model
log_reg = LogisticRegression(max_iter=1000, random_state=123)

# Cross-validated predictions
y_pred_log = cross_val_predict(log_reg, X_train_smote, y_train_smote, cv=5)

print("📊 Logistic Regression Classification Report:\n")
print(classification_report(y_train_smote, y_pred_log))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest
rf = RandomForestClassifier(random_state=123)
y_pred_rf = cross_val_predict(rf, X_train_smote, y_train_smote, cv=5)

print("🌲 Random Forest Classification Report:\n")
print(classification_report(y_train_smote, y_pred_rf))
```

## Gradient Boost

```python
from sklearn.ensemble import GradientBoostingClassifier

# Gradient Boosting
gb = GradientBoostingClassifier(random_state=123)
y_pred_gb = cross_val_predict(gb, X_train_smote, y_train_smote, cv=5)

print("🔥 Gradient Boosting Classification Report:\n")
print(classification_report(y_train_smote, y_pred_gb))
```

# ROC-AUC Curve

```python
from sklearn.metrics import roc_curve, auc

# Model fitting
log_reg.fit(X_train_smote, y_train_smote)
rf.fit(X_train_smote, y_train_smote)
gb.fit(X_train_smote, y_train_smote)

# Probabilities for ROC
y_prob_log = log_reg.predict_proba(X_test)[:, 1]
y_prob_rf = rf.predict_proba(X_test)[:, 1]
y_prob_gb = gb.predict_proba(X_test)[:, 1]

# ROC curves
fpr_log, tpr_log, _ = roc_curve(y_test, y_prob_log)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
fpr_gb, tpr_gb, _ = roc_curve(y_test, y_prob_gb)

# AUC scores
auc_log = auc(fpr_log, tpr_log)
auc_rf = auc(fpr_rf, tpr_rf)
auc_gb = auc(fpr_gb, tpr_gb)


plt.figure(figsize=(10, 6))
plt.plot(fpr_log, tpr_log, label=f'Logistic Regression (AUC = {auc_log:.2f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.2f})')
plt.plot(fpr_gb, tpr_gb, label=f'Gradient Boosting (AUC = {auc_gb:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')

plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

y_test_pred_log = log_reg.predict(X_test)
y_test_pred_rf = rf.predict(X_test)
y_test_pred_gb = gb.predict(X_test)

# Confusion Matrices
print("📘 Logistic Regression Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_log))
print("\n🌲 Random Forest Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_rf))
print("\n🔥 Gradient Boosting Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_gb))
```

```python
# In employee turnover prediction, our goal is to catch "true leavers" so that HR can intervene.

#Recall (Sensitivity) is critical:

#“Of all employees who were actually going to leave, how many did we correctly predict?”

#Precision matters too, but lower recall would mean we miss people who are at risk.

# So our priority metric is: Recall for class 1 (leavers)
```

# Turnover probabilities 


```python
# Predicting the probability of class 1 (employee leaving)
y_test_probs = rf.predict_proba(X_test)[:, 1]


prob_df = pd.DataFrame({
    'Actual Left': y_test,
    'Probability of Leaving': y_test_probs
})

def categorize_risk(score):
    if score < 0.20:
        return 'Safe Zone (Green)'
    elif score < 0.60:
        return 'Low-Risk Zone (Yellow)'
    elif score < 0.90:
        return 'Medium-Risk Zone (Orange)'
    else:
        return 'High-Risk Zone (Red)'

prob_df['Risk Zone'] = prob_df['Probability of Leaving'].apply(categorize_risk)

# Count employees in each zone
zone_counts = prob_df['Risk Zone'].value_counts()
print("📊 Employee Count by Risk Zone:\n")
print(zone_counts)
```

# Creating a pie to plot

```python

# Counting the risk zones
zone_counts = prob_df['Risk Zone'].value_counts()


colors = {
    'Safe Zone (Green)': '#4CAF50',
    'Low-Risk Zone (Yellow)': '#FFEB3B',
    'Medium-Risk Zone (Orange)': '#FF9800',
    'High-Risk Zone (Red)': '#F44336'
}


plt.figure(figsize=(4, 4))
plt.pie(zone_counts, 
        labels=zone_counts.index,
        colors=[colors[label] for label in zone_counts.index],
        autopct='%1.1f%%',
        startangle=140,
        wedgeprops=dict(width=0.4))  # for a donut look

plt.title('📊 Employee Turnover Risk Distribution', fontsize=14)
plt.axis('equal')  # Equal aspect ratio makes the pie round
plt.show()
```
